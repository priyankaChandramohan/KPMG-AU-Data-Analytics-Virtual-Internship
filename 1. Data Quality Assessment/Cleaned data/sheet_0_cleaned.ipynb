```python
import pandas as pd
import numpy as np
```

```python
#reading the excel file and converting to dataframe
df3 = pd.read_excel('C:/Users/Priyanka C/Desktop/KPMG Virtual Internship/KPMG_VI_New_raw_data_update_final.xlsx', sheet_name=0, header = 0)
```

```python
df3.info()
```

```python
df3.head()
```

```python
#sorting values based on customer id and transaction date
df3 = df3.sort_values(['customer_id', 'transaction_date'])
```

```python
#count of online orders
df3['online_order'].value_counts()
```

```python
#filling null values in online order
df3['online_order'] = df3['online_order'].fillna(method = 'bfill')
```

```python
df3.info()
```

```python
#filling null values in brand
df3['brand'] = df3['brand'].fillna(method = 'ffill')
```

```python
#filling null values in product line
df3['product_line'] = df3['product_line'].fillna(method = 'bfill')
```

```python
#filling null values in product class
df3['product_class'] = df3['product_class'].fillna(method = 'bfill')
```

```python
#filling null values in product size
df3['product_size'] = df3['product_size'].fillna(method = 'ffill')
```

```python
df3.info()
```

```python
# fill the missing data with the mean costs for that customer
df3['standard_cost'] = df3['standard_cost'].fillna(df3['standard_cost'].mean())
```

```python
# fill the missing data with the mean for product first sold date
df3['product_first_sold_date'] = df3['product_first_sold_date'].fillna(df3['product_first_sold_date'].mean())
```

```python
#checking data after cleaning
df3.info()
```
